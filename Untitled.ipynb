```python
import numpy as np
import cv2

import coco
import segmentacion as seg
import datos_nombres as dt
```

```python
img_train=[]
img_train_dot=[]
img_test=[]
labels_test={}
etiquetas_color={
    "0":[[0, 0, 100], [50, 50, 255], 20, 0.00000159,0.000000579],  #Correlaciones max:2.7411419244361876e-06, min:7.382692675463257e-07, media:1.8857256819759879e-06 Correlaciones max:8.741839516943169e-07, min:3.24245104897081e-07, media:6.64751713520656e-07
    "1":[[200, 0, 200], [255, 100, 255],15,0.000000812,0.000000442], #Correlaciones max:1.1905213390456026e-06, min:5.496838268757373e-07, media:9.882204942233206e-07  Correlaciones max:5.773732037863762e-07, min:3.4094390826560933e-07, media:5.100231243555509e-07
    "2":[[15, 25, 105], [40, 70, 150],16,0.00000145,0.000000735], #Correlaciones max:1.5847357146872362e-06, min:4.5296136394543194e-07, media:1.1689069326786435e-06  Correlaciones max:7.418906537075732e-07, min:1.9581862779708544e-07, media:5.59946153195829e-07 
    "3":[[151, 48, 24], [180, 60, 50],10,0.000000135,0.000000135],
    "4":[[20, 160, 30], [40, 180, 60],7,0.00000022,0.1] #Correlaciones max:2.586514118829518e-07, min:8.576618900881228e-08, media:2.4334599107835095e-07
}
coordenadas={
}
prototipos={}
```

```python
#Obtener img de entrenamiento redimensionadas
for nombre in dt.TRAIN:
    print(nombre)
    ruta=f"img/Train/{nombre}.jpg"
    ruta_dot=f"img/TrainDotted/{nombre}.jpg"
    img_train.append(seg.redim(ruta,2000))
    img_train_dot.append(seg.redim(ruta_dot,2000))

#Obtener img de prueba redimensionadas y sus resultados
for nombre in dt.TEST:
    print(nombre)
    ruta=f"img/Train/{nombre}.jpg"
    img_test.append(seg.redim(ruta,2000))
    labels_test.update(dt.obtener_etiquetas(nombre,"img/Train/train.csv"))
```

```python
#Obtener los prototipos ENTRENAMIENTO
croco_proto=10 #Correlacin entre prototipos
for clase in range(5):
    clave=f"{clase}"
    print(clave)
    lista_prototipos=list()
    taman_proto=1+etiquetas_color[clave][2]*2
    for img in range(len(img_train)):
        encontrados = seg.encontrar_puntos_color(img_train_dot[img],etiquetas_color[clave][0],etiquetas_color[clave][1])
        for (y, x) in encontrados:
            prototipo=seg.obtener_subarreglo(img_train[img],y,x,etiquetas_color[clave][2])
            if len(lista_prototipos)==0:
                lista_prototipos.append(prototipo)
            else:
                if (prototipo.shape)[0]!=taman_proto or (prototipo.shape)[1]!=taman_proto:
                    prototipo=lista_prototipos[0]
                else:
                    lista_prototipos=coco.actualizar_lista_prototipos(lista_prototipos,prototipo,croco_proto)
    prototipos[clave]=lista_prototipos
```

```python
#Buscar leones en las imagenes nuevas
R=0.00000068 #Porcentaje de correlacin para decidir clase
for img in range(len(img_test)):
    print(f"Buscando en img test {img}")
    img_etiquetada=np.copy(img_test[img])
    for clase in range(3,4):
        print(f"Clase: {clase}")
        coord_clase=[]
        clave=f"{clase}"
        R=etiquetas_color[clave][3+img]
        i=0
        lista_corr_max=[]
        for proto in prototipos[clave]:
            i+=1
            print(f"Protipo {i}/{len(prototipos[clave])}")
            coord_clase.extend(coco.croco_buscar_nuevos(proto, etiquetas_color[clave][2], img_test[img], R,lista_corr_max))
            #print(coord_clase)
        print("Filtrando coordenadas de clase")
        coordenadas_ordenadas = sorted(coord_clase, key=lambda x: (x[0], x[1]))
        coord_clase=coco.filtrar_coordenadas(coordenadas_ordenadas, etiquetas_color[clave][2])
        print(f"Colocando etiquetas {len(coord_clase)}")
        print(f"Correlaciones max:{max(lista_corr_max)}, min:{min(lista_corr_max)}, media:{sum(lista_corr_max) / len(lista_corr_max)}")
        img_etiquetada=seg.agregar_cuadros_envolventes(coord_clase, img_etiquetada, etiquetas_color[clave][1], etiquetas_color[clave][2])
    # Guardar la imagen con los cuadros envolventes
    cv2.imwrite(f"imagen_con_cuadros{img}.png", img_etiquetada)
```
